Este proyecto propone el desarrollo de un sistema de detección temprana de enfermedades mediante técnicas de aprendizaje automático, utilizando datos obtenidos de exámenes sanguíneos y parámetros clínicos básicos, como niveles de glucosa, presión sanguínea y otras variables relevantes.

A partir de estos datos, se construyen y evalúan modelos de aprendizaje supervisado (como regresión múltiple y árboles de decisión) con el objetivo de identificar patrones asociados a la presencia o riesgo de determinadas enfermedades. El enfoque se centra en analizar la capacidad predictiva de los modelos y su utilidad como herramienta de apoyo al diagnóstico clínico, sin sustituir la evaluación médica profesional.

Los resultados del estudio buscan demostrar cómo el uso de técnicas de aprendizaje automático puede contribuir a una detección más temprana y eficiente de posibles enfermedades, facilitando la toma de decisiones en contextos clínicos y resaltando el potencial de estas metodologías en el análisis de datos médicos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import dataset
data = pd.read_csv('data/diabetes.csv')
```

```python
print(data.head())
print(data.describe())
# Usando seaborn para visualizar la distribución de las características
sns.pairplot(data, hue='Outcome')

```

```python
# Matris de correlación
corr = data.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriz de Correlación')
```

```python
# Histograma de (Glucosa, BMI, Edad) para ver su distribución (Con Diabetes y Sin Diabetes)
features = ['Glucose', 'BMI', 'Age']    
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, feature in enumerate(features):
    sns.histplot(data, x=feature, hue='Outcome', bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribución de {feature} por Resultado de Diabetes')
plt.tight_layout()
plt.show()
```
